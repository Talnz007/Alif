```python
import re
```

```python
def clean_text(text: str) -> str:
    """Normalize text: lowercase, remove punctuation, and unwanted characters"""
    # Lowercase the text
    text = text.lower()
    
    # Remove punctuation and digits
    text = re.sub(r'[^\w\s]', '', text)
    
    # Remove extra spaces and newlines
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text
```

```python
text = "Apple Inc. is a technology company based in Cupertino, California, founded in 1976."
cleaned_text = clean_text(text)
cleaned_text
```
